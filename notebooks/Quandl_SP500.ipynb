```python
import pandas as pd
import numpy as np
```

```python
path_ticker = '/Users/ralf/Documents/6000_stocks/s_and_p_500/SHARADAR_TICKERS_6cc728d11002ab9cb99aa8654a6b9f4e.csv'
path_fundamentals = '/Users/ralf/Documents/6000_stocks/s_and_p_500/SHARADAR_SF1_086134faf658fcc2cdcb53f4295b5fad.csv'
df_fundamentals = pd.read_csv(path_fundamentals)
df_tickers = pd.read_csv(path_ticker)
```

```python
df_fundamentals[df_fundamentals["ticker"] == "SAP"]
```

```python
df_fundamentals["yyy"] = df_fundamentals["datekey"].str.split("-", expand = True)[0]
```

```python
df_fundamentals["yyy"] = df_fundamentals["yyy"].astype(int)
```

```python
list(df_fundamentals["yyy"] .unique())
```

```python
df_selection = df_fundamentals[['ticker', 'dimension','calendardate','netinc','assets','debt','ncfdiv',
                                 'shareswa', 'sharesbas', 'marketcap', 'pe']]
ind_anual = df_selection['dimension'] == 'MRY'
df_selection_anual = df_selection[ind_anual]
```

```python
ind_date = df_selection_anual['calendardate'] == '2018-12-31'
df_2018 = df_selection_anual[ind_date]
```

```python
df_2018['ROTA'] = df_2018['netinc']/df_2018['assets']*100
#df_2018['earningsyield'] = df_2018['netinc']/df_2018['marketcap']*100
df_2018['earningsyield'] = 1/df_2018['pe']*100


df_2018_filtered = df_2018[['ticker','netinc','assets','debt','ncfdiv','marketcap','earningsyield', 'ROTA']]
ind_yield = df_2018_filtered['earningsyield'] > 2.0
df_2018_filtered_2 = df_2018_filtered[ind_yield]
ind_rota = df_2018_filtered_2['ROTA'] > 9.0
df_2018_filtered_3 = df_2018_filtered_2[ind_rota]
ind_marketcap = df_2018_filtered_3['marketcap'] > 3e+09
df_2018_filtered_4 = df_2018_filtered_3[ind_marketcap]
df_2018_filtered_4.count()
```

```python
df_t = df_tickers[['ticker', 'name', 'industry']]
df_merged = pd.merge(df_2018_filtered_4, df_t, how = 'left',left_on = 'ticker', right_on = 'ticker')
df_merged_2 = df_merged.drop_duplicates().sort_values(by=['ROTA','earningsyield'], ascending = False)
df_merged_2['netinc'] = df_merged_2['netinc'] /1e+09
df_merged_2['assets'] = df_merged_2['assets'] /1e+09
df_merged_2['debt'] = df_merged_2['debt'] /1e+09
df_merged_2['ncfdiv'] = df_merged_2['ncfdiv'] /1e+09
df_merged_2['marketcap'] = df_merged_2['marketcap'] /1e+09
```
